```python
%%writefile topsis/topsis.py
import sys
import pandas as pd
import numpy as np

def topsis(input_file, weights, impacts, output_file):
    df = pd.read_csv(input_file)
    matrix = df.iloc[:, 1:].values.astype(float)

    weights = list(map(float, weights.split(',')))
    impacts = impacts.split(',')

    norm = matrix / np.sqrt((matrix ** 2).sum(axis=0))
    weighted = norm * weights

    best, worst = [], []
    for i in range(len(impacts)):
        if impacts[i] == '+':
            best.append(weighted[:, i].max())
            worst.append(weighted[:, i].min())
        else:
            best.append(weighted[:, i].min())
            worst.append(weighted[:, i].max())

    d_plus = np.sqrt(((weighted - best) ** 2).sum(axis=1))
    d_minus = np.sqrt(((weighted - worst) ** 2).sum(axis=1))

    score = d_minus / (d_plus + d_minus)
    df["Topsis Score"] = score
    df["Rank"] = df["Topsis Score"].rank(ascending=False)

    df.to_csv(output_file, index=False)

def main():
    topsis(sys.argv[1], sys.argv[2], sys.argv[3], sys.argv[4])
```
